```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Start coding!
netflix_df = pd.read_csv('data/netflix_data.csv')
netflix_df.head()
```

```python
# Subsetting the DataFrame to include only rows where the type is 'Movie'
netflix_subset = netflix_df[netflix_df['type'] == 'Movie']
```

```python
netflix_movies = netflix_subset[['title', 'country', 'genre', 'release_year', 'duration']]
```

```python
short_movies = netflix_movies[netflix_movies.duration < 60]
```

```python
# Initialize an empty list called colors
colors = []

# Iterate through netflix_movies rows
for index, row in netflix_movies.iterrows():
    # Check if the genre is "Children", "Documentaries", or "Stand-Up"
    if row['genre'] == 'Children':
        colors.append('red')
    elif row['genre'] == 'Documentaries':
        colors.append('blue')
    elif row['genre'] == 'Stand-Up':
        colors.append('green')
    else:
        colors.append('black')
```

```python
# Inspect the first 10 values in your list        
colors[:10]
```

```python
fig = plt.figure(figsize=(12, 8))
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)
plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.title('Movie Duration by Year of Release')
plt.show()
```

```python
#"Are we certain that movies are getting shorter?
answer = "no"
```

```python
#netflix_movies.country = netflix_movies.country.apply(lambda x: 'USA' if x.upper() == 'UNITED STATES' else x)
netflix_movies.duration = netflix_movies.duration.str.rstrip('min')
netflix_movies.duration = pd.to_numeric(netflix_movies.duration)
netflix_movies.release_year = netflix_movies.release_year.astype(int)

# Group by the release year and calculate the average duration of movies
short_movies = netflix_movies[netflix_movies.duration < 60]
short_movies.release_year.value_counts().sort_index()
short_movies.release_year.value_counts().sort_index().plot(kind='bar', figsize=(12, 8))
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies Released by Year')
plt.show()
```
